# Inflammation code

The input file is a set of numbers where each row represents a patient, each column is a day and each item is a level of inflammation (0,20). Days are presented consecutevely. It is a CSV file.

We want to find out the average inflammation of **all** the patients per day.

## Reading the file

The data is stored in a list of lists. Each inner list at position _i_ contains the readings of inflammation on the _i_-th day for all patients.

```python
file = open("inflammation/inflammation-all.csv")

ds = []
all = []
for i in range(40):
    ds.append([])
    
for line in file:
    vals = line.strip().split(',')
    for i in range(40):
        ds[i].append(int(vals[i]))
        all.append(int(vals[i]))
        
print("Max inflammation:", max(all))
print("Min inflammation:", min(all))
```

## Inflammation evolution

The first graph presented shows the evolution of inflammation throughout all 40 days. The green line represents the maximum inflamation of the day. The yellow line is the minimum inflammation of the day. The red line is the average.

```python
import matplotlib.pyplot as plt

x = range(1,41)
y = []
max_y = []
min_y = []

for l in ds:
    y.append(sum(l) / len(l))
    max_y.append(max(l))
    min_y.append(min(l))
    
plt.plot(x, y, 'r', x, max_y, 'g', x, min_y, 'y')
plt.title("Inflammation x days")
plt.xlabel("Days")
plt.ylabel("Inflammation level")
plt.show()
```

## Analysis

The peak of inflammation is at around 20 days.

## Histograms

We will know look at histograms of the inflammation level per day to see how this evolves.

```python
plt.figure(figsize=(20,35))

for i in range(40):
    plt.subplot(8,5,i+1)
    infl_on_day = ds[i]
    plt.title("Day " + str(i+1))
    plt.hist(infl_on_day, range(0,21,2), rwidth=0.5)
    
plt.show()
```
