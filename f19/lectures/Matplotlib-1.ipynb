```python
import matplotlib as mpl
import matplotlib.pyplot as plt  # import the module
mpl.rcParams['figure.dpi']= 130  # set the resolution to x dpi
```

```python
plt.figure()  # create a new figure, not strictly necessary

plt.title("My first graph, only dots")  # the title to give to the figure
plt.xlabel("x axis")  # x-axis label
plt.ylabel("y axis")  # y-axis label

plt.plot(1, 2, marker='p', color='red', markersize=10)

plt.plot(10, 1, marker='x', markersize=10)

plt.show()
```

```python
plt.figure()  # create a new figure, not strictly necessary

plt.title("My first graph, only dots")  # the title to give to the figure
plt.xlabel("x axis")  # x-axis label
plt.ylabel("y axis")  # y-axis label

# each command plots an (x,y) point with a color and a marker
#
plt.plot(1, 1, marker='.', color='blue')
plt.plot(2, 4, marker='o', color='red')

plt.plot(3, 9, 'b.')  # short form, b stands for blue, . is marker
plt.plot(4, 16,'ro')  # short form, r stands for red, o is marker
plt.plot(25,'g+')

yy = [2,5,7,11,13]
plt.plot(yy,'rx')

xx = [0,1,2,3,4]
plt.plot(xx, yy,'rx')


plt.plot(5, 25, marker='x', color='green')
plt.plot(5, 25, marker='x', color='green')
plt.plot(6, 36, marker='p', color='yellow')

plt.show() # show in the figure all the issued plot commands

#plt.plot([1,2,3,4,5],[1,4,9,16,25], color="red")
#plt.show()
```

```python
import numpy as np  # import numpy to use arange and math functions

plt.figure(figsize=(5, 3))  # create a new figure with a given size

plt.title("Plot (x,y) sequences with lines")  
plt.xlabel("x")  
plt.ylabel("y")  

x = np.arange(-5, 6, 1)  # 11 x coordinate points
y = np.arange(2, 3.1, 0.1)  # 11 y coordinate points

#plt.plot(x, y, marker='.', color='blue') 
plt.plot(x,y)
# with lists, the default behavior is to plot with connecting lines

plt.show()
```

```python
plt.figure(figsize=(2, 3))  # create a new figure with a given size

plt.title("How to control x,y limits and tick marks\n")  
plt.xlabel("x")  
plt.ylabel("y")  

x = range(-5, 5, 1)  # 10 x coordinate points
plt.xlim(-5, 5) # limits for the x axis

y = np.arange(2.5, 3.5, 0.1)
plt.ylim(2, 5) # limits for the y axis
plt.yticks(np.arange(2, 5.5, 0.5))  # where to place the tick marks on y axis

plt.plot(x, y, marker='.', color='red') 

plt.show()
```

```python
import random as rnd

plt.figure(figsize=(5, 3))  # create a new figure with a given size

plt.title("Plot randomly generated NASDAQ data, use textual labels\n", fontsize=10)  
plt.xlabel("Month (Year 2050)", fontsize=9)  
plt.ylabel("NASDAQ", fontsize=9)  

# we want to plot a randomly generated value for NASDAQ for each month of the year
x = range(1, 13, 1)  # 12 months in a year

# a list of texual labels to place at the tick marks on the x axis
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 
          'Sep', 'Oct', 'Nov', 'Dec']

# the labels in list months are places at the positions defined by the list x
plt.xticks(x, months, fontsize=8)

# let's give a reasonable range for nasdaq values
nasdaq_min = 10000
nasdaq_max = 15000

# generate 12 random values for the NASDAQ one per each month, between min and max 
y = []
for n in range(12):
    y.append(rnd.uniform(nasdaq_min, nasdaq_max))
    # the uniform(a,b) method from random module generates a random number 
    # uniformly selected in the real interval range [a,b]
 
'''
#this is another way yo do it using the method sample() from module random
list_to_sample_from = list(np.arange(nasdaq_min, nasdaq_max, 0.25))
num_samples = 12
y = rnd.sample( list_to_sample_from, num_samples)
'''

# set the limits on the y axis based on the min and max values used for sampling
#plt.ylim(nasdaq_min, nasdaq_max)
#plt.ylim(12000, 13000)
# set the tick marcks for y axis every 500
plt.yticks(range(nasdaq_min, nasdaq_max+1, 500), fontsize=8)

plt.plot(x, y, color='red', marker='d', linestyle='--', linewidth=1) 

plt.show()
```

```python
plt.figure(figsize=(2, 3))  

plt.title("Plot a function y=f(x)")  
plt.xlabel("x")  
plt.ylabel("$y = x^2$")  

xmin = -1
xmax = 1
xstep = 0.1
x = np.arange(xmin, xmax + xstep, xstep)  #  x coordinate points
y = []
for xval in x:
    y.append(xval**2)

plt.plot(x, y)
```

```python
plt.title("Plot multiple functions\n", fontsize=10)  
plt.xlabel("x")  
plt.ylabel("$y = f(x)$")  

x = np.arange(-1, 1.1, 0.01)  #  x coordinate points

y = []
for xval in x:
    y.append(xval**3)  # y coordinates, y = x^2
plt.plot(x, y, color = 'red')

y  = []
for xval in x:
    y.append(np.cos(6 * xval))  # y coordinates, y = x^2
plt.plot(x, y, color = 'green')
```

```python
plt.figure()

plt.title("Plot multiple functions with legends\n", fontsize=10)  
plt.xlabel("x")  
plt.ylabel("$y = f(x)$")  

x = np.arange(-1, 1.1, 0.01)  #  x coordinate points

y2 = []
for xval in x:
    y2.append((xval)**3)  # y coordinates, y = x^2
    
y3  = []
for xval in x:
    y3.append(np.cos(6 * xval))  # y coordinates, y = x^2

y4  = []
for xval in x:
    y4.append(np.sqrt(1+xval))  # y coordinates, y = x^2

# add legends to the plot
y2_legend, = plt.plot(x, y2, label='$x^3$')
y3_legend, = plt.plot(x, y3, label='$\cos(6x)$')
y4_legend, = plt.plot(x, y4, label='$\sqrt{1+x}$')

# plot the legends with the labels
plt.legend(handles=[y2_legend, y3_legend, y4_legend])
plt.show()
```
