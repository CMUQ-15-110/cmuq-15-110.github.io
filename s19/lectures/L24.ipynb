```python
import random
def roll_die():
    return random.choice([1,2,3,4,5,6])
```

```python
def pascal_problem(num_trials=10000):
    both_six = 0
    for t in range(num_trials):
        for i in range(24):
            die_1 = roll_die()
            die_2 = roll_die()
            if die_1 == 6 and die_2 == 6:
                both_six += 1
                break
    return both_six / num_trials
```

***
The number of trials equal the number of **simulated experiments** that we will try out to construct the *empirical probability* for not getting a double 6 after rolling the dice 24 times
***

```python
prob = pascal_problem(10000)
print('Combinatorial probability of losing the game: {:4.3f}'.format((35/36) ** 24))
print('Empirical probability of losing the game: {:4.3f}'.format(1-prob))
```

***
What if **dice aren't fair?** Let's put more weight on the 6 face of a dice
***

```python
def roll_die_unfair():
    return random.choice([1,2,3,4,5,6,6])
```

```python
# Let's use a bit more flexible function definition.
# The function now takes as input two function arguments, one per each die roll
#
def pascal_problem(num_trials, roll_1, roll_2):
    both_six = 0
    for t in range(num_trials):
        for i in range(24):
            die_1 = roll_1()
            die_2 = roll_2()
            if die_1 == 6 and die_2 == 6:
                both_six += 1
                break
    return both_six / num_trials
```

***
What if *both dice* are unfair?
***

```python
prob = pascal_problem(20000, roll_die_unfair, roll_die_unfair)
print('Combinatorial probability of losing the game: {:4.3f}'.format((35/36) ** 24))
print('Empirical probability of losing the game: {:4.3f}'.format(1-prob))
```

***
What if only *one die* is unfair?
***

```python
prob = pascal_problem(20000, roll_die, roll_die_unfair)
print('Combinatorial probability of losing the game: {:4.3f}'.format((35/36) ** 24))
print('Empirical probability of losing the game: {:4.3f}'.format(1-prob))
```

***
A simple **linear congruential generator (LCG)** of random numbers (bits, 0/1, head/tails)
***

```python
def LCG(state, a=1664525, c=1013904223, m=2**32-1):
    next_state = (a * state + c) % m
    return (next_state)

state = 1
for i in range(100):
    state = LCG(state)
    print(state % 2, '', end='')
```

Use and effects of the `seed(val=Null)` method

```python
seed = 12356
print('First call of PRNG, seed = ', seed)
random.seed(seed)
for i in range(100):
    print(random.randint(0,1), '', end='')

random.seed(seed)
print('\n')
print('Second call of PRNG, seed = ', seed)
for i in range(100):
    print(random.randint(0,1), '', end='')
print('\n')

seed = 12357
random.seed(seed)
print('Third call of PRNG, seed = ', seed)
for i in range(100):
    print(random.randint(0,1), '', end='')
print('\n')
```

```python
import matplotlib.pyplot as plt

sequence = []
seed = 98765
random.seed(seed)
for i in range(100):
    sequence.append(random.randint(0,1))

plt.figure()
plt.title('A random sequence of 0 and 1')
plt.xlabel('Trial')
plt.ylabel('Outcome')
plt.plot(sequence)
plt.show()
```

```python
sequences = []
num_sequences = 4
for t in range(num_sequences):
    seed = (t+1) * 98765
    random.seed(seed)
    sequences.append([])
    for i in range(100):
        sequences[t].append(random.randint(0,1))

plt.figure(figsize = (8,6))
plt.title('Random sequences of 0 and 1')
plt.xlabel('Trial')
plt.ylabel('Outcome')
for i in range(num_sequences):
    plt.plot(sequences[i])
plt.show()
```

```python
sequences = []
num_sequences = 2
for t in range(num_sequences):
    seed = (t+1) * 98765
    random.seed(seed)
    sequences.append([])
    for i in range(100):
        sequences[t].append(random.randint(0,1))

plt.figure(figsize = (7,5))
plt.title('Random sequences of 0 and 1')
plt.xlabel('Trial')
plt.ylabel('Outcome')
for i in range(num_sequences):
    #
    # RGB color encoding: a triple (r,g,b), where r,g,b are values in [0,1]
    # and represent respectively the amount of red (r), green (g), blue (b)
    # in the final color blend (e.g., (1,0,0) would produce a red color)
    #
    rgb = (random.uniform(0,1), random.uniform(0,1), random.uniform(0,1))
    plt.plot(sequences[i], linestyle='None', marker='.', markersize=5, color=rgb)
plt.show()
```
